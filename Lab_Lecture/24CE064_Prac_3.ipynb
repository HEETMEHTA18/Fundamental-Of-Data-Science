```python

import pandas as pd
import numpy as np

students = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4],
    'Name': ['Amit', 'Riya', 'Sohan', 'Neha']
})

marks = pd.DataFrame({
    'Student_ID': [1, 2, 3],
    'Marks': [85, 90, 78]
})

print("Students:\n", students)
print("\nMarks:\n", marks)
```

```python
inner_join = pd.merge(students, marks, on='Student_ID', how='inner')
print("\nInner Join:\n", inner_join)

left_join = pd.merge(students, marks, on='Student_ID', how='left')
print("\nLeft Join:\n", left_join)

right_join = pd.merge(students, marks, on='Student_ID', how='right')
print("\nRight Join:\n", right_join)

outer_join = pd.merge(students, marks, on='Student_ID', how='outer')
print("\nOuter Join:\n", outer_join)

```

```python
students_indexed = students.set_index('Student_ID')
marks_indexed = marks.set_index('Student_ID')

index_merge = pd.merge(
    students_indexed,
    marks_indexed,
    left_index=True,
    right_index=True,
    how='inner'
)

print("\nIndex-Based Merge:\n", index_merge)
```

```python
sales = pd.DataFrame({
    'Product_ID': [101, 102, 103],
    'Quantity': [5, 3, 10]
})

price = pd.DataFrame({
    'Product_ID': [101, 102, 103],
    'Price': [200, 150, 50]
})

merged_sales = pd.merge(sales, price, on='Product_ID')

merged_sales['Total_Cost'] = np.multiply(
    merged_sales['Quantity'],
    merged_sales['Price']
)

print("\nSales with Total Cost:\n", merged_sales)
```

```python
city = ['Ahmedabad', 'Ahmedabad', 'Surat', 'Surat', 'Surat']
weather = ['Sunny', 'Rainy', 'Sunny', 'Cloudy', 'Rainy']

crosstab = pd.crosstab(city, weather)
print(crosstab)
```

```python
percentage = np.round(
    crosstab.div(crosstab.sum(axis=1), axis=0) * 100,
    2
)
print(percentage)
```

```python
import pandas as pd
import numpy as np


jan_orders = pd.DataFrame({
    'Order_ID': [1, 2, 3],
    'Product_ID': [101, 102, 101],
    'Quantity': [2, 1, 3],
    'Month': 'Jan',
    'Payment_Type': ['UPI', 'Card', 'UPI']
})


feb_orders = pd.DataFrame({
    'Order_ID': [4, 5],
    'Product_ID': [102, 103],
    'Quantity': [2, 5],
    'Month': 'Feb',
    'Payment_Type': ['Cash', 'UPI']
})
```

```python
orders = pd.concat([jan_orders, feb_orders], ignore_index=True)
print(orders)
```

```python
products = pd.DataFrame({
    'Product_ID': [101, 102, 103],
    'Product_Name': ['Notebook', 'Pen', 'Bag'],
    'Price': [50, 20, 300]
})
```

```python
merged_data = pd.merge(orders, products, on='Product_ID', how='left')
print(merged_data)
```

```python
merged_data['Total_Sales'] = np.multiply(
    merged_data['Quantity'],
    merged_data['Price']
)
print(merged_data)
```

```python
pivot_sales = pd.pivot_table(
    merged_data,
    index='Month',
    columns='Product_Name',
    values='Total_Sales',
    aggfunc='sum',
    fill_value=0
)

print(pivot_sales)
```

```python
payment_crosstab = pd.crosstab(
    merged_data['Month'],
    merged_data['Payment_Type']
)

print(payment_crosstab)
```

```python
product_sales = merged_data.groupby('Product_Name')['Total_Sales'].sum()

percentage_contribution = np.round(
    (product_sales / product_sales.sum()) * 100,
    2
)

print(percentage_contribution)
```

```python
import pandas as pd
import numpy as np

q1 = pd.DataFrame({
    'Product_ID': [101, 102, 101],
    'Quantity': [10, 5, 8],
    'Quarter': 'Q1'
})

q2 = pd.DataFrame({
    'Product_ID': [101, 103],
    'Quantity': [7, 12],
    'Quarter': 'Q2'
})
q3 = pd.DataFrame({
    'Product_ID': [102, 103],
    'Quantity': [9, 6],
    'Quarter': 'Q3'
})
```

```python
annual_sales = pd.concat([q1, q2, q3], ignore_index=True)
print("Annual Data:\n", annual_sales)
```

```python
prices = pd.DataFrame({
    'Product_ID': [101, 102, 103],
    'Price': [50, 20, 300]
})
```

```python
merged_sales = pd.merge(annual_sales, prices, on='Product_ID', how='left')
print("\nMerged Data:\n", merged_sales)
```

```python
merged_sales['Revenue'] = np.multiply(
    merged_sales['Quantity'],
    merged_sales['Price']
)

print("\n Revenue:\n", merged_sales)
```

```python
total_revenue = merged_sales.groupby('Product_ID')['Revenue'].sum()
print("\nTotal Revenue:\n", total_revenue)
```

```python
import pandas as pd
import numpy as np

data = {
    'Department': ['CE', 'CE', 'IT', 'IT', 'CE', 'IT', 'CE'],
    'Grade': ['A', 'B', 'A', 'B', 'A', 'A', 'B']
}

df = pd.DataFrame(data)
print(df)
```

```python
ct = pd.crosstab(df['Department'], df['Grade'])
print("\nCrosstab \n", ct)
```

```python
percentage = np.round(
    (ct / ct.sum(axis=1).values.reshape(-1, 1)) * 100,2)
print("\nPercentage Distribution:\n", percentage)
```

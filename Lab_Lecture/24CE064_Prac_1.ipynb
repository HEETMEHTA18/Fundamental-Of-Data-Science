```python
# 1 Create one-dimensional NumPy arrays
import numpy as np
import pandas as pd

arr_int = np.array([1,20,3,4,5])
print("integer array",arr_int)

arr_float = np.array([1,2.1,3.4,2.0])
print("integer array",arr_float)

arr_mixed = np.array([1,2,'two',True])
print("integer array",arr_mixed)
```

```python
print("\nAddition",arr_int+10)
print("Subtraction:", arr_int - 2)
print("Multiplication:", arr_int * 3)
print("Division:", arr_int / 2)

print("\nMean:", np.mean(arr_int))
print("Max:", np.max(arr_int))
print("Min:", np.min(arr_int))
print("Standard Deviation:", np.std(arr_int))
```

```python
import pandas as pd
import numpy as np


df = pd.read_csv("indata.csv")   
df
```

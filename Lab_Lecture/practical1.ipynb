```python
import numpy as np
import pandas as pd
```

```python
print("NumPy Version:", np.__version__)
print("Pandas Version:", pd.__version__)
```

```python
arr_int = np.array([10, 20, 30, 40, 50])
arr_float = np.array([1.5, 2.5, 3.5, 4.5])
arr_mixed = np.array([10, 2.5, "Hello"])
```

```python
print("Integer Array:", arr_int)
print("Float Array:", arr_float)
print("Mixed Array:", arr_mixed)
```

```python
x = np.array([10, 20, 30, 40])
y = np.array([2, 4, 6, 8])

print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
```

```python
print("Mean:", np.mean(x))
print("Max:", np.max(x))
print("Min:", np.min(x))
print("Standard Deviation:", np.std(x))
```

```python
sales = np.array([1200, 1500, 1800, 1100, 1600, 2000, 1700])
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

s = pd.Series(sales, index=days)
print(s)
```

```python
print("First element using iloc:", s.iloc[0])
print("Slice using iloc [1:4]:\n", s.iloc[1:4])
```

```python
print("Value of Mon using loc:", s.loc["Mon"])
print("Mon to Thu using loc:\n", s.loc["Mon":"Thu"])
```

```python
print("Head:\n", s.head())
print("Tail:\n", s.tail())
print("Describe:\n", s.describe())
```

```python
data = {
    "Name": ["Aman", "Riya", "Hetu", "Neha"],
    "Math": [85, 92, 78, 88],
    "Science": [80, 90, 75, 95]
}

df = pd.DataFrame(data)
df
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Index:", df.index)
print("Dtypes:\n", df.dtypes)

df.info()
```

```python
df.describe()
```

```python
high_math = df[df["Math"] > 85]
high_math
```

```python
df["Total"] = df["Math"] + df["Science"]
df
```

```python
df["Percentage"] = (df["Total"] / 200) * 100
df
```

```python
df.to_csv("student_result.csv", index=False)
print("Saved as student_result.csv")
```

```python
df_new = pd.read_csv("student_result.csv")
print("Shape after reload:", df_new.shape)
df_new.head()
```

```python
numeric_cols = df_new.select_dtypes(include=np.number)

print("Numeric columns:\n", numeric_cols)

print("\nColumn-wise Mean (NumPy):\n", np.mean(numeric_cols, axis=0))
print("\nColumn-wise Median (NumPy):\n", np.median(numeric_cols, axis=0))
```

```python
row_totals = numeric_cols.sum(axis=1)
max_index = row_totals.idxmax()

print("Row with highest total:\n")
df_new.loc[max_index]
```

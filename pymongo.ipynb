```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client['mydatabase']
collection = db['mycollection']

collection.insert_one({"name": "Alice", "age": 30})
collection.insert_one({"name": "Bob", "age": 25})
```

collection.insert_one("")

```python
collection.insert_one({"name":'Heet',"age":'20',"marks":'00',"address":{'city':'chonga','state':'guj','pincode':'388421'}})
collection.insert_many([{"name":'bob','age':25,'marks':85,'address':{'city':'surat','state':'guj','pincode':'395001'}
                         },{"name":'alice','age':30,'marks':90,'address':{'city':'vadodara','state':'guj','pincode':'390001'}}
                         ,{"name":'charlie','age':35,'marks':80,'address':{'city':'ahmedabad','state':'guj','pincode':'380001'}}

                         ])
```

```python
collection.find_one({"name": "Alice"})
collection.find({"name": "Alice"})
```

```python
collection.find()
collection.find_one()
```

```python
for i in collection.find({},{ "_id": 0,"name": 1,"age":1,"marks":1,"address":1}):  #in this the projection is done like sql and first is make as empty
    print(i)
```

```python
#filtration
for i in collection.find({"age": {"$lt": 50}}):
    print(i)
# collection.find({"age": {"$gt": 25}})
```

```python
#filter with projecection
for doc in collection.find({},{"_id": 0, "name": 1, "age": 1, "marks": 1, "address.state": 1}):
    print(doc)
```

```python
for doc in collection.find({},{'_id':0,'age':26}):
    print(doc)
```

```python
#cursoer maethod
for dov in collection.find({}, {"_id": 0}).sort("age", -1).limit(1):
    print(dov)

for doc in collection.find({}, {"_id": 0}).sort("age", 1).limit(10).skip(1):
    print(doc)
```

```python
# this is the implicit data typecasting in which the datatype is changed to the float 
import numpy as np
a= np.array([10,20.5,30])
b= np.array([1,2,3.0])
print(a+b)
print(a.dtype)
```

```python
# print("min:",np.sub(a,b) there is no function for the addition and subtraction 
print("mean",np.mean(a))
print("mean",np.mean(b))
print("std dev:",np.std(a))
print("max:",np.max(a))
print("max:",np.max(b))
print("min:",np.min(a))
print("min:",np.min(b))
print(a.dtype)
print(b.dtype)
```

```python
days= ["Mon","Tue","Wed","Thru","Fri","Sat","Sun"]
temps = ([30, 40, 50, 12, 35, 25, 10]) 

seq =pd.Series(temps , index =days) 
print(seq)
print(seq.iloc[1])
print(seq['Mon'])
print(seq[::3])
print(seq[:'Tue'])

print(seq.head())
print("first 3",seq.head(3))


print("-------------",seq.head())
print("-------------",seq.tail())
```

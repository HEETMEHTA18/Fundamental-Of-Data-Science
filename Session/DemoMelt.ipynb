```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv")
```

```python
df
```

```python
melted = pd.melt(df, id_vars=["day"], var_name='city', value_name='temperature')
melted
```

```python
pivot_df = melted.pivot(index='day',
                           columns='city',
                           values='temperature')
```

```python
import pandas as pd 
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
pd.merge(employees,departments,on="DeptID")  #Inner Join And Which work on common DepartID
```

```python
pd.merge(employees,departments,on="DeptID",how = "left") # here left means that left side emplyoees 
                                                         # so aena badha coloum condier thay
```

```python
pd.merge(employees,departments,on="DeptID",how ="right")#right side ma department che to aena 
                                                        #badha colum consider thay 
```

```python
pd.merge(employees,departments,on="DeptID",how ="outer")# badhi row and Coloum consider thay
```
